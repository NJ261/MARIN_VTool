### Read input data
#### Here, distance matrix is generate based on one-to-six points calculation. With the one-to-all points, it wasn't fulfilling triangle inequality hence resulting paths weren't optimized paths.
#### Further, data contains calculations for 10x10 sample grids.
#### Distance unit in csv file is meter.

```python
csvToMatrix = CSVtoMatrix('../Data/DistanceMatrix/oneToSixPoints_10x10.csv')
inputData = csvToMatrix.inputData()
inputData.head()
```

### Preparing graph for dijkstra
#### Here, we will prepare graph which contains node to node distance

```python
graph = Graph()

for i in range(0, len(inputData)):
    graph.addEdge(inputData['InputID'][i], inputData['TargetID'][i], inputData['Distance'][i])
```

### Result
#### The output will give node id from origin to destination

```python
dijkstra = Djikstra(graph, 18425, 17376)
path = dijkstra.djikstra()
print(path)
```
